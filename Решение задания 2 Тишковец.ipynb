```python
branch_Vor = df1.groupby("Категория")['Воронеж на Цирке'].apply(lambda x: x.sum()).to_numpy() # переоценок по филиалу 6719
branch_Eis = df1.groupby("Категория")['Ейск ТЦ ВАП'].apply(lambda x: x.sum()).to_numpy() # переоценок по филиалу 5178
branch_Samara = df1.groupby("Категория")['Самара ТЦ Космопорт Гипер'].apply(lambda x: x.sum()).to_numpy() # переоценок по филиалу 8711
branch_Tum = df1.groupby("Категория")['Тюмень Ямская Гипер'].apply(lambda x: x.sum()).to_numpy() # переоценок по филиалу  6289
total_sum = branch_Vor[0]+branch_Eis[0]+branch_Samara[0]+branch_Tum[0]
# пусть сотрудники работают по 10 часов 15 дней в месяц, тогда в час он получает 333.3 рубля
wage_per_hour = 50000 / 150 
cost_Vor1 = round(branch_Vor[0]*2/60*wage_per_hour,1)
cost_Eis = round(branch_Eis[0]*2/60*wage_per_hour,1)
cost_Samara = round(branch_Samara[0]*2/60*wage_per_hour,1)
cost_Tum = round(branch_Tum[0]*2/60*wage_per_hour,1)
cost_total = round((cost_Vor1 + cost_Eis + cost_Samara + cost_Tum)/4)
print('количество переоценок за январь Воронеж на Цирке =', branch_Vor[0], '\n'
     'количество переоценок за январь Ейск ТЦ ВАП =', branch_Eis[0], '\n'
     'количество переоценок за январь Самара ТЦ Космопорт Гипер =', branch_Samara[0], '\n'
     'количество переоценок за январь Тюмень Ямская Гипер =', branch_Tum[0],'\n'
     'общее для всех филиалов количество переоценок за январь  =', total_sum, '\n'
     'стоимость процесса «Переоценка» Воронеж на Цирке за январь  =', cost_Vor1, 'руб' , '\n'
     'стоимость процесса «Переоценка» Ейск ТЦ ВАП за январь  =', cost_Eis, 'руб' , '\n'
     'стоимость процесса «Переоценка» Самара ТЦ Космопорт Гипер за январь  =', cost_Samara, 'руб' , '\n'
     'стоимость процесса «Переоценка» Тюмень Ямская Гипер за январь  =', cost_Tum, 'руб' , '\n'
     'средняя стоимость процесса «Переоценка» для всех филиалов за январь  =', cost_total, 'руб' , '\n')
```
